```python
class gen_PkRatio():
    def __init__(self,data_dir,zbin_logPkR,scenario,z,kh):
        self.data_dir    = data_dir
        self.zbin_logPkR = zbin_logPkR
        self.scenario = scenario
        self.infile_logPkRatio = self.data_dir + "logPkRatio_" + scenario + ".dat"
        #self.DF_logPkR         = self.read_logPkR_file()
        self.Bins_z            = self.zbin_logPkR[scenario]
        self.f_logPkRatio      = f_logPkRatio
    
    def read_logPkR_file(self):
        DF_logPkR = pd.read_csv(self.infile_logPkRatio,sep='\s+')
        return DF_logPkR
    
    def f_logPkRatio(self):
        Arr2D_logPkratio = self.read_logPkR_file()
        Bins_log_k_Mpc = np.array(Arr2D_logPkratio["logk"])
        del Arr2D_logPkratio["logk"]
        Arr2D_logPkratio = np.array(Arr2D_logPkratio)
        f_logPkRatio = interpolate.interp2d(self.Bins_z, Bins_log_k_Mpc, Arr2D_logPkratio, kind='linear')
        return f_logPkRatio
```
